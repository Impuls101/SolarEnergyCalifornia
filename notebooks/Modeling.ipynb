## Data Modeling
This notebook includes the data modeling steps in the solar energy project. <br>
As inputs it requires the cleaned, encoded, re-sampled, and scaled data frames X_train, X_test, y_train, y_test. <br>
The objective is to test different models

### 0. Data collection

```python
import pandas as pd
```

```python
path = r"..\data\processed\\" # for Manuel

X_train_scaled_df = pd.read_csv(path+'X_train.csv')
X_test_scaled_df = pd.read_csv(path+'X_test.csv')
y_train_resampled_cc = pd.read_csv(path+'y_train.csv')
y_test = pd.read_csv(path+'y_test.csv')
```

```python
X_train_scaled_df.head(1)
```

### 1. Random Forest Classifier

```python
# Only test-wise I used a baseline model (RandomForestClassifier) to see if the pre-processing was okay (so this notebook cell is just a test):
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Instantiating the model (e.g., RandomForestClassifier) and train the model on the scaled training data (y_train does not need to be scaled since it is binary)
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled_df, y_train_resampled_cc)

# Create predictions
y_pred = model.predict(X_test_scaled_df)

# Evaluate model performance using a classification report
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```

### 2. SVM Classifier

```python
from sklearn.svm import SVC

# Instantiate the model and train it
svm_model = SVC(kernel='rbf', probability=True, random_state=42)
svm_model.fit(X_train_scaled_df, y_train_resampled_cc)

# Create predictions
svm_pred = svm_model.predict(X_test_scaled_df)

# Evaluate model performance using a classification report
svm_report = classification_report(y_test, svm_pred)
print("Classification Report:")
print(svm_report)
```

### 3. XGBoost Classifier

```python
from xgboost import XGBClassifier

# Instantiate the model and train it
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(X_train_scaled_df, y_train_resampled_cc)

# Create predictions
xgb_pred = xgb_model.predict(X_test_scaled_df)

# Evaluate model performance using a classification report
xgb_report = classification_report(y_test, xgb_pred)
print("Classification Report:")
print(xgb_report)
```

### 4. K-Nearest Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the model and train it
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_scaled_df, y_train_resampled_cc)

# Create predictions
knn_pred = knn_model.predict(X_test_scaled_df)

# Evaluate model performance using a classification report
knn_report = classification_report(y_test, knn_pred)
print("Classification Report:")
print(knn_report)
```
